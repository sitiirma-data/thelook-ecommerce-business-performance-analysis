Sementara itu, variabel `age` memiliki rentang nilai antara 12 hingga 70 tahun, yang masih berada dalam rentang usia yang wajar untuk pelanggan e-commerce. Namun demikian, keberadaan outlier tidak dapat ditentukan hanya berdasarkan statistik deskriptif sehingga diperlukan pemeriksaan lanjutan menggunakan boxplot dan metode Interquartile Range (IQR).

```python
#Visualisasi Boxplot
import matplotlib.pyplot as plt

columns_to_check = {
    "Users": (users, ["age"]),
    "Products": (products, ["cost", "retail_price"]),
    "Inventory Items": (inventory, ["cost", "product_retail_price"]),
    "Order Items": (order_items, ["sale_price"])
}

for table_name, (df, cols) in columns_to_check.items():
    for col in cols:
        plt.figure(figsize=(6, 1.8))
        plt.boxplot(df[col].dropna(), vert=False)
        plt.title(f"{table_name} - {col}")
        plt.show()
```

**Interpretasi**

Berdasarkan visualisasi boxplot, variabel **cost**, **retail_price**, **product_retail_price**, dan **sale_price** menunjukkan adanya sejumlah nilai yang berada di luar batas atas (upper whisker) sehingga secara statistik teridentifikasi sebagai outlier. Hal ini menunjukkan bahwa terdapat produk dengan harga atau biaya yang jauh lebih tinggi dibandingkan mayoritas produk lainnya.

Sementara itu, variabel **age** tidak menunjukkan adanya outlier. Rentang usia pelanggan masih berada dalam batas distribusi yang wajar sehingga tidak memerlukan penanganan lebih lanjut.

Keberadaan outlier pada variabel harga dan biaya belum dapat langsung dianggap sebagai kesalahan data. Mengingat dataset berasal dari perusahaan e-commerce dengan beragam kategori produk, nilai ekstrem tersebut berpotensi merepresentasikan produk premium atau produk dengan harga tinggi. Oleh karena itu, diperlukan validasi kuantitatif menggunakan metode Interquartile Range (IQR) sebelum menentukan apakah outlier perlu ditangani pada tahap data cleaning.

```python
#Menghitung jumlah Outlier (Metode IQR)
def count_outliers_iqr(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    return ((series < lower) | (series > upper)).sum()

outlier_summary = []

for table_name, (df, cols) in columns_to_check.items():
    for col in cols:
        outlier_summary.append({
            "Dataset": table_name,
            "Column": col,
            "Outlier Count": count_outliers_iqr(df[col].dropna())
        })

pd.DataFrame(outlier_summary)
```

**Interpretasi**

Berdasarkan metode Interquartile Range (IQR), tidak ditemukan outlier pada variabel `age`. Hal ini menunjukkan bahwa distribusi usia pelanggan masih berada dalam rentang yang wajar.

Sebaliknya, variabel yang berkaitan dengan harga dan biaya (`cost`, `retail_price`, `product_retail_price`, dan `sale_price`) memiliki sejumlah nilai yang teridentifikasi sebagai outlier. Hasil ini konsisten dengan visualisasi boxplot yang menunjukkan adanya nilai ekstrem pada batas atas distribusi.

Meskipun demikian, outlier tersebut tidak dihapus dari dataset. Nilai-nilai ekstrem masih berada dalam rentang harga yang realistis untuk produk e-commerce dan kemungkinan merepresentasikan produk premium atau produk dengan nilai jual tinggi, sehingga dianggap sebagai variasi bisnis yang valid, bukan kesalahan pencatatan data.

Oleh karena itu, seluruh outlier dipertahankan agar informasi penting mengenai variasi harga produk tetap dapat dimanfaatkan pada tahap analisis berikutnya.

Berdasarkan hasil Data Quality Assessment, ditemukan bahwa beberapa kolom bertipe tanggal masih tersimpan sebagai `object`. Selain itu, hasil pemeriksaan konsistensi format tanggal menunjukkan adanya variasi format timestamp, yaitu sebagian data memiliki presisi hingga detik, sedangkan sebagian lainnya hingga microsecond.

Oleh karena itu, dilakukan konversi seluruh kolom tanggal menjadi tipe data `datetime` menggunakan parameter `format="mixed"` agar seluruh variasi format timestamp dapat diproses secara konsisten.

```python
date_columns = {
    "users": ["created_at"],
    "orders": ["created_at", "shipped_at", "delivered_at", "returned_at"],
    "order_items": ["created_at", "shipped_at", "delivered_at", "returned_at"],
    "inventory_items": ["created_at", "sold_at"],
    "events": ["created_at"]
}

for dataset_name, columns in date_columns.items():
    for col in columns:
        datasets[dataset_name][col] = pd.to_datetime(
            datasets[dataset_name][col],
            format="mixed"
        )
```

#### Interpretasi

Proses konversi tipe data berhasil dilakukan pada seluruh kolom yang merepresentasikan informasi tanggal dan waktu. Konversi dilakukan menggunakan `pd.to_datetime()` dengan parameter `format="mixed"` untuk mengakomodasi variasi format timestamp yang ditemukan pada tahap Data Quality Assessment.

Melalui proses ini, seluruh kolom tanggal telah dipersiapkan agar dapat digunakan pada analisis berbasis waktu, seperti analisis tren penjualan, durasi pengiriman, dan perilaku pelanggan.

#### 4.2.2 Validasi Hasil Konversi

Setelah proses konversi selesai dilakukan, dilakukan validasi kembali terhadap tipe data setiap tabel untuk memastikan seluruh kolom tanggal telah berhasil dikonversi menjadi tipe data `datetime`.

#### 5.1.3 Categorical Analysis

Analisis variabel kategorikal dilakukan untuk memberikan gambaran mengenai karakteristik variabel non-numerik pada dataset. Analisis ini mencakup jumlah kategori yang dimiliki setiap variabel, kategori yang paling dominan, serta proporsi kemunculannya. Informasi ini digunakan untuk memahami komposisi data sebelum dilakukan analisis hubungan antar variabel pada tahap berikutnya.

```python
categorical_summary = []

categorical_columns = {
    "Users": (users, ["gender", "traffic_source", "country"]),
    "Orders": (orders, ["status"]),
    "Products": (products, ["category", "brand"]),
    "Events": (events, ["event_type", "browser"])
}

for dataset_name, (df, columns) in categorical_columns.items():
    for col in columns:

        dominant = df[col].value_counts().idxmax()
        frequency = df[col].value_counts().max()
        percentage = (frequency / len(df)) * 100

        categorical_summary.append({
            "Dataset": dataset_name,
            "Variable": col,
            "Number of Categories": df[col].nunique(),
            "Dominant Category": dominant,
            "Frequency": frequency,
            "Percentage (%)": round(percentage,2)
        })

categorical_summary = pd.DataFrame(categorical_summary)

categorical_summary
```

Performa bisnis dianalisis melalui tren **revenue**, **jumlah transaksi (total orders)**, **jumlah pelanggan aktif (active customers)**, dan **Average Order Value (AOV)** untuk memahami apakah pertumbuhan bisnis lebih dipengaruhi oleh peningkatan volume transaksi, pertumbuhan pelanggan, atau perubahan nilai transaksi rata-rata.

**Data Preparation**

Sebelum melakukan analisis, data transaksi diaggregasi pada tingkat bulanan untuk membangun metrik utama yang merepresentasikan performa bisnis. Metrik yang digunakan meliputi total revenue, jumlah transaksi, jumlah pelanggan aktif, dan Average Order Value (AOV). Seluruh metrik dihitung menggunakan periode waktu yang sama agar hasil analisis konsisten dan dapat dibandingkan secara langsung.

```python
# ==========================================================
# Data Preparation - Business Performance Summary
# ==========================================================

# Merge Order Items dengan Orders
business_df = (
    order_items
    .drop(columns=["created_at"])
    .merge(
        orders[["order_id", "created_at"]],
        on="order_id",
        how="left"
    )
)

# Membuat periode bulanan
business_df["year_month"] = (
    business_df["created_at"]
    .dt.tz_localize(None)
    .dt.to_period("M")
    .dt.to_timestamp()
)

# Business Performance Summary
business_performance = (
    business_df
    .groupby("year_month", as_index=False)
    .agg(
        Revenue=("sale_price", "sum"),
        Total_Orders=("order_id", "nunique"),
        Active_Customers=("user_id", "nunique")
    )
)

# Average Order Value (AOV)
business_performance["AOV"] = (
    business_performance["Revenue"] /
    business_performance["Total_Orders"]
)

# Menampilkan hasil
business_performance.head()
```

####5.2.2 Product Performance Analysis


**Business Question 2**

Karakteristik produk seperti apa yang memberikan kontribusi terbesar terhadap pertumbuhan bisnis, dan faktor apa yang membedakan produk berkinerja tinggi dengan produk berkinerja rendah?



**Tujuan Analisis**

Analisis ini bertujuan untuk mengidentifikasi karakteristik produk yang memberikan kontribusi terbesar terhadap performa bisnis TheLook. Evaluasi dilakukan dengan menganalisis kontribusi setiap kategori produk berdasarkan revenue, jumlah produk terjual (quantity sold), profit, dan profit margin untuk memahami faktor-faktor yang membedakan kategori produk berkinerja tinggi dan berkinerja rendah. Hasil analisis ini diharapkan dapat membantu perusahaan mengidentifikasi kategori produk yang layak diprioritaskan dalam strategi pengembangan bisnis.



**Data Preparation**

Untuk mendukung analisis performa produk, data transaksi pada tabel `order_items` digabungkan dengan informasi produk pada tabel `products` berdasarkan `product_id`. Penggabungan ini bertujuan untuk menghubungkan data penjualan dengan karakteristik setiap produk sehingga performa masing-masing kategori dapat dievaluasi secara menyeluruh.

Selanjutnya dihitung beberapa metrik utama yang digunakan dalam analisis, yaitu **Revenue**, **Quantity Sold**, **Profit**, dan **Profit Margin**. Seluruh metrik kemudian diagregasi pada tingkat kategori produk sehingga setiap kategori dapat dibandingkan secara konsisten dalam mengevaluasi kontribusinya terhadap performa bisnis.

```python
# Menggabungkan data transaksi dengan data produk
product_performance = (
    order_items.merge(
        products[
            [
                "id",
                "category",
                "brand",
                "cost",
                "retail_price"
            ]
        ],
        left_on="product_id",
        right_on="id",
        how="left"
    )
)

# Menghitung profit setiap transaksi
product_performance["profit"] = (
    product_performance["sale_price"]
    - product_performance["cost"]
)

product_performance.head()
```

```python
product_performance.shape
```

```python
# Agregasi performa berdasarkan kategori produk
category_performance = (
    product_performance
    .groupby("category", as_index=False)
    .agg(
        Revenue=("sale_price", "sum"),
        Quantity_Sold=("id_x", "count"),
        Profit=("profit", "sum")
    )
)

# Menghitung Profit Margin (%)
category_performance["Profit_Margin"] = (
    category_performance["Profit"]
    / category_performance["Revenue"]
    * 100
).round(2)

# Mengurutkan berdasarkan Revenue terbesar
category_performance = (
    category_performance
    .sort_values(
        by="Revenue",
        ascending=False
    )
    .reset_index(drop=True)
)

category_performance
```

#### Interpretasi

Hasil agregasi menunjukkan bahwa setiap kategori produk memiliki karakteristik performa yang berbeda berdasarkan empat metrik utama, yaitu Revenue, Quantity Sold, Profit, dan Profit Margin. Perbedaan tersebut mengindikasikan bahwa kategori produk tidak hanya dapat dievaluasi dari besarnya revenue, tetapi juga dari jumlah produk yang terjual, keuntungan yang dihasilkan, serta efisiensi dalam menghasilkan profit.

Namun, pada tahap ini belum dapat disimpulkan kategori produk mana yang memiliki performa terbaik karena analisis masih berupa hasil agregasi. Oleh karena itu, visualisasi diperlukan untuk membandingkan setiap metrik secara lebih jelas dan mengidentifikasi karakteristik kategori produk berkinerja tinggi maupun berkinerja rendah.

####5.2.3 Customer Behavior Analysis

Bagaimana perbedaan perilaku pembelian pelanggan berdasarkan frekuensi transaksi, dan bagaimana pengaruhnya terhadap performa bisnis TheLook?

**Tujuan Analisis**

Analisis ini bertujuan untuk mengevaluasi perbedaan perilaku pembelian pelanggan berdasarkan frekuensi transaksi serta memahami pengaruhnya terhadap performa bisnis TheLook. Analisis dilakukan dengan membandingkan frekuensi transaksi, kontribusi revenue, dan karakteristik pelanggan untuk mengidentifikasi pola pembelian yang memberikan nilai bisnis lebih tinggi serta menjadi dasar dalam penyusunan strategi customer retention.


**Data Preparation**

Sebelum melakukan analisis, data transaksi diagregasi pada tingkat pelanggan (**customer-level aggregation**) untuk membangun metrik utama yang merepresentasikan perilaku pembelian setiap pelanggan.

Metrik yang digunakan meliputi:

- **Transaction Frequency**, yaitu jumlah transaksi yang dilakukan setiap pelanggan.
- **Total Revenue**, yaitu total revenue yang dihasilkan setiap pelanggan.
- **Average Order Value (AOV)**, yaitu rata-rata nilai transaksi setiap pelanggan.

Selain itu, karakteristik pelanggan seperti **Age Group**, **Gender**, **Country**, dan **Traffic Source** ditambahkan untuk mendukung analisis karakteristik pelanggan berdasarkan frekuensi transaksi.

```python
users.info()
```

```python
# Mengagregasi data transaksi pada tingkat customer
# untuk menghitung frekuensi transaksi, total revenue,
# dan Average Order Value (AOV)
customer_performance = (
    order_items
    .groupby("user_id", as_index=False)
    .agg(
        Transaction_Frequency=("order_id", "nunique"),
        Total_Revenue=("sale_price", "sum"),
        AOV=("sale_price", "mean")
    )
)

# Menggabungkan hasil agregasi dengan data customer
# untuk menambahkan karakteristik pelanggan
customer_performance = (
    customer_performance
    .merge(
        users[
            [
                "id",
                "age",
                "age_group",
                "gender",
                "country",
                "traffic_source"
            ]
        ],
        left_on="user_id",
        right_on="id",
        how="left"
    )
)

# Menampilkan hasil agregasi customer
customer_performance.head()
```

```python
# Menghapus kolom id karena sudah direpresentasikan oleh user_id
customer_performance.drop(columns="id", inplace=True)

customer_performance.head()
```

```python
# Melihat distribusi frekuensi transaksi setiap customer
transaction_distribution = (
    customer_performance["Transaction_Frequency"]
    .value_counts()
    .sort_index()
)

transaction_distribution
```

#### 5.2.4 Operational Performance Analysis

**Business Question 4**

Pada tahapan proses operasional mana perusahaan kehilangan peluang bisnis terbesar, dan faktor apa yang berkontribusi terhadap kondisi tersebut?

**Tujuan Analisis**

Analisis ini bertujuan untuk mengidentifikasi tahapan proses operasional yang menyebabkan perusahaan kehilangan peluang bisnis terbesar selama periode observasi. Evaluasi dilakukan melalui analisis distribusi status pesanan, jumlah pesanan, serta revenue pada setiap tahapan proses operasional untuk mengetahui sumber utama kehilangan peluang bisnis dan faktor-faktor yang berkontribusi terhadap kondisi tersebut.

**Data Preparation**

Sebelum melakukan analisis, data transaksi diagregasi berdasarkan **Order Status** untuk membangun metrik utama yang merepresentasikan performa operasional bisnis.

Metrik yang digunakan meliputi:

- **Total Orders**, yaitu jumlah pesanan pada setiap status.
- **Revenue**, yaitu total revenue yang terkait dengan masing-masing status pesanan.
- **Order Percentage**, yaitu persentase jumlah pesanan pada setiap status terhadap keseluruhan pesanan.

Agregasi ini digunakan sebagai dasar untuk mengidentifikasi tahapan operasional yang berpotensi menyebabkan kehilangan peluang bisnis serta mengevaluasi kontribusi masing-masing status terhadap performa operasional perusahaan.

```python
order_items["status"].value_counts()
```

```python
# Mengagregasi data transaksi berdasarkan status pesanan
# untuk menghitung jumlah pesanan, total revenue,
# dan rata-rata nilai transaksi (Average Order Value)
operational_performance = (
    order_items
    .groupby("status", as_index=False)
    .agg(
        Total_Orders=("order_id", "nunique"),
        Revenue=("sale_price", "sum"),
        Average_Order_Value=("sale_price", "mean")
    )
)

# Menghitung persentase jumlah pesanan
# pada setiap status terhadap total keseluruhan pesanan
operational_performance["Order_Percentage"] = (
    operational_performance["Total_Orders"]
    / operational_performance["Total_Orders"].sum()
    * 100
).round(2)

# Menampilkan hasil agregasi performa operasional
operational_performance
```

#### Interpretasi

#### 5.2.5 Traffic Source Performance Analysis

**Business Question 5**

Bagaimana efektivitas setiap sumber traffic dalam menghasilkan pelanggan yang memiliki nilai bisnis tinggi serta memberikan kontribusi terhadap performa bisnis perusahaan?

**Tujuan Analisis**

Analisis ini bertujuan untuk mengevaluasi efektivitas setiap **Traffic Source** dalam menghasilkan pelanggan yang memberikan nilai bisnis tinggi serta kontribusi terhadap performa bisnis perusahaan. Evaluasi dilakukan dengan membandingkan jumlah pelanggan, total revenue, rata-rata revenue per pelanggan, frekuensi transaksi, dan Average Order Value (AOV) pada masing-masing sumber traffic untuk mengidentifikasi saluran akuisisi pelanggan yang paling efektif.

**Data Preparation**

Sebelum melakukan analisis, data transaksi diagregasi berdasarkan **Traffic Source** untuk membangun metrik utama yang merepresentasikan performa setiap sumber traffic.

Metrik yang digunakan meliputi:

- **Total Customers**, yaitu jumlah pelanggan unik yang berasal dari setiap sumber traffic.
- **Total Revenue**, yaitu total revenue yang dihasilkan oleh pelanggan pada setiap sumber traffic.
- **Average Revenue per Customer**, yaitu rata-rata revenue yang dihasilkan setiap pelanggan.
- **Average Transaction Frequency**, yaitu rata-rata frekuensi transaksi pelanggan pada setiap sumber traffic.
- **Average Order Value (AOV)**, yaitu rata-rata nilai transaksi pelanggan pada setiap sumber traffic.

Agregasi ini digunakan untuk membandingkan efektivitas setiap sumber traffic dalam menghasilkan pelanggan bernilai tinggi serta kontribusinya terhadap performa bisnis perusahaan.

```python
# Mengagregasi performa pelanggan berdasarkan sumber traffic
traffic_performance = (
    customer_performance
    .groupby("traffic_source", as_index=False)
    .agg(
        Total_Customers=("user_id", "nunique"),
        Total_Revenue=("Total_Revenue", "sum"),
        Average_Revenue_Per_Customer=("Total_Revenue", "mean"),
        Average_Transaction_Frequency=("Transaction_Frequency", "mean"),
        Average_Order_Value=("AOV", "mean")
    )
)

# Mengurutkan data berdasarkan total revenue terbesar
traffic_performance = (
    traffic_performance
    .sort_values(
        by="Total_Revenue",
        ascending=False
    )
)

# Menampilkan hasil agregasi performa setiap sumber traffic
traffic_performance
```

#### Interpretasi

Data transaksi berhasil diagregasi berdasarkan **Traffic Source** sehingga diperoleh metrik utama berupa **Total Customers**, **Total Revenue**, **Average Revenue per Customer**, **Average Transaction Frequency**, dan **Average Order Value (AOV)** untuk setiap sumber traffic.

Hasil agregasi menunjukkan bahwa **Search** memiliki jumlah pelanggan dan total revenue yang jauh lebih tinggi dibandingkan sumber traffic lainnya. Namun, nilai **Average Revenue per Customer**, **Average Transaction Frequency**, dan **Average Order Value (AOV)** relatif serupa pada seluruh sumber traffic. Hal ini mengindikasikan bahwa perbedaan performa antar traffic source kemungkinan lebih dipengaruhi oleh **kemampuan mengakuisisi pelanggan dalam jumlah besar** daripada perbedaan kualitas pelanggan yang dihasilkan.

Oleh karena itu, analisis selanjutnya akan difokuskan pada perbandingan kontribusi revenue dan jumlah pelanggan dari setiap sumber traffic untuk mengidentifikasi saluran akuisisi yang paling efektif dalam mendukung performa bisnis perusahaan.

## Export Dataset for Power BI

Pada tahap ini dilakukan penggabungan data transaksi, produk, dan pelanggan ke dalam satu master dataset yang akan digunakan sebagai sumber data pada dashboard Power BI.

Dataset ini telah mencakup seluruh atribut yang diperlukan untuk visualisasi sehingga proses analisis pada Power BI dapat dilakukan tanpa perlu melakukan proses join ulang.

```python
# Menggabungkan data transaksi dengan data produk dan data pelanggan
powerbi_dataset = (
    order_items
    .merge(
        products[
            [
                "id",
                "category",
                "brand",
                "cost",
                "retail_price"
            ]
        ],
        left_on="product_id",
        right_on="id",
        how="left"
    )
    .merge(
        users[
            [
                "id",
                "age",
                "age_group",
                "gender",
                "country",
                "traffic_source"
            ]
        ],
        left_on="user_id",
        right_on="id",
        how="left",
        suffixes=("", "_user")
    )
)

# Menghitung profit setiap transaksi
powerbi_dataset["profit"] = (
    powerbi_dataset["sale_price"]
    - powerbi_dataset["cost"]
)

# Menampilkan dataset final
powerbi_dataset.head()
```

```python
# Mengekspor dataset final untuk Power BI
powerbi_dataset.to_csv(
    "thelook_powerbi_dataset.csv",
    index=False
)
```
